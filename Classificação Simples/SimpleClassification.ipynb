```python
#Eliminate unique/or almost unique columns
df_final=df.drop(['id', 'poi'], axis=1)
df_final
```

```python
#Simple Label encoding
#Import label encoder
from sklearn import preprocessing
  
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'category'.
df_final['category']= label_encoder.fit_transform(df_final['category'])
  
df_final['category'].unique()
```

```python
# Encode labels in column 'transport_mode'.
df_final['transport_mode']= label_encoder.fit_transform(df_final['transport_mode'])
  
df_final['transport_mode'].unique()
```

```python
#1º Classification problem - transport mode
Y1=df_final['transport_mode'].to_numpy()  #1st possible dependent variable
type(Y1)
```

```python
#2nd Classification Problem - purpose
Y2=df_final['purpose'].to_numpy()  #2nd possible dependent variable
```

```python
#Dependent variables are removed - 1st Problme
X1=df_final.drop(['transport_mode','purpose'], axis=1).to_numpy()
print(X1)
```

```python
#Dependent variable is remove - 1st problem
X2=df_final.drop('purpose', axis=1).to_numpy()
print(X2)
```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split
# Split dataset into training set and test set
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, Y1, test_size=0.3) # 70% training and 30% test
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, Y2, test_size=0.3) # 70% training and 30% test
```

```python
#Training and testing a decicion tree classifier - transport mode
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

clf1 = DecisionTreeClassifier().fit(X1_train, y1_train)
y1_pred=clf1.predict(X1_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y1_test, y1_pred))
print("F1-Score:",metrics.f1_score(y1_test, y1_pred, average='macro'))
print("Confusion Matrix:\n", metrics.confusion_matrix(y1_test, y1_pred))
```
