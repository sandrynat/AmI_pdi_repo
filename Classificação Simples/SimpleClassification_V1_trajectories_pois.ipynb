```python
from collections import Counter
# Function to replace the label with the most repeated value
def replace_with_most_repeated(label):
    # Split the string into a list
    activities = label.split(';')
    # Find the most common activity
    most_common_activity = Counter(activities).most_common(1)[0][0]
    return most_common_activity

# Apply the function to the 'label' column
df['label'] = df['label'].apply(replace_with_most_repeated)
df['activity_label'] = df['activity_label'].apply(replace_with_most_repeated)
```

```python
#Eliminate unique/or almost unique columns
df=df.drop([ 'uid1', 'trajectory_id'], axis=1)
df
```

```python
# Fill NaN values with 'Missing'
df['category'].fillna('Missing', inplace=True)
print("\nDataFrame after handling NaN values:")
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False)
encoded_data = encoder.fit_transform(df[['category']])
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['category']))
print("Encoded DataFrame:")
print(encoded_df)
```

```python
df = pd.concat([df, encoded_df], axis=1)
df
```

```python
#handle datetime
df['start_time'] = pd.to_datetime(df['start_time'])
def get_day_type(date):
    return 'WEEKEND' if date.weekday() >= 5 else 'WEEKDAY'

def get_time_of_day(date):
    hour = date.hour
    if 5 <= hour < 12:
        return 'MORNING'
    elif 12 <= hour < 17:
        return 'AFTERNOON'
    elif 17 <= hour < 21:
        return 'EVENING'
    else:
        return 'NIGHT'
df['DayType'] = df['start_time'].apply(get_day_type)
df['TimeOfDay'] = df['start_time'].apply(get_time_of_day)
```

```python
df.info()
df
```

```python
#one hot encoding DayType
encoded_data1 = encoder.fit_transform(df[['DayType']])
encoded_df1 = pd.DataFrame(encoded_data1, columns=encoder.get_feature_names_out(['DayType']))
df = pd.concat([df, encoded_df1], axis=1)

#one hot encoding TimeOfDay
encoded_data2 = encoder.fit_transform(df[['TimeOfDay']])
encoded_df2 = pd.DataFrame(encoded_data2, columns=encoder.get_feature_names_out(['TimeOfDay']))
df = pd.concat([df, encoded_df2], axis=1)

df
```

```python
#remove categorical and datetime columns
#Eliminate unique/or almost unique columns
df_final=df
df_final=df_final.drop([ 'TimeOfDay', 'DayType','start_time', 'category', 'duration'], axis=1)
```

```python
#Simple Label encoding
#Import label encoder


from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'activity_label'.
df_final['activity_label']= label_encoder.fit_transform(df_final['activity_label'])

df_final['activity_label'].unique()
```

```python
# Encode labels in column 'label'.
df_final['label']= label_encoder.fit_transform(df_final['label'])

df_final['label'].unique()
```

```python
#1º Classification problem - transport mode
Y1=df_final['activity_label'].to_numpy()  #1st possible dependent variable
type(Y1)
```

```python
#2nd Classification Problem - purpose
Y2=df_final['label'].to_numpy()  #2nd possible dependent variable
```

```python
#Dependent variables are removed - 1st Problme
X1=df_final.drop(['activity_label','label'], axis=1).to_numpy()
print(X1)
```

```python
#Dependent variable is remove - 1st problem
X2=df_final.drop('label', axis=1).to_numpy()
print(X2)
```

```python
df_final.info()
```

```python
#min-max normalization
import pandas as pd
# Min-Max Normalization function
def min_max_normalize(df2):
    return (df2 - df2.min()) / (df2.max() - df2.min())

# Apply Min-Max Normalization
X2 = min_max_normalize(X2)
X2
```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split
# Split dataset into training set and test set
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, Y1, test_size=0.3,random_state=123) # 70% training and 30% test
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, Y2, test_size=0.3,random_state=123) # 70% training and 30% test
```

```python
#Training and testing a decicion tree classifier - transport mode
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

clf1 = DecisionTreeClassifier(random_state=76).fit(X1_train, y1_train)
y1_pred=clf1.predict(X1_test)

# Model Accuracy, how often is the classifier correct?
print("transport mode")
print("Accuracy:",metrics.accuracy_score(y1_test, y1_pred))
print("F1-Score:",metrics.f1_score(y1_test, y1_pred, average='macro'))
print("Confusion Matrix:\n", metrics.confusion_matrix(y1_test, y1_pred))


clf2 = DecisionTreeClassifier(random_state=77).fit(X2_train, y2_train)
y2_pred=clf2.predict(X2_test)
print("label")
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y2_test, y2_pred))
print("F1-Score:",metrics.f1_score(y2_test, y2_pred, average='macro'))
print("Confusion Matrix:\n", metrics.confusion_matrix(y2_test, y2_pred))


import matplotlib.pyplot as plt
import seaborn as sns

# Compute confusion matrix
cm = metrics.confusion_matrix(y1_test, y1_pred)

# Define the color map
cmap = sns.light_palette("blue", as_cmap=True)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap=cmap, fmt='g', cbar=False)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix activity')
plt.show()

# Compute confusion matrix
cm2 = metrics.confusion_matrix(y2_test, y2_pred)

# Define the color map
cmap2 = sns.light_palette("blue", as_cmap=True)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm2, annot=True, cmap=cmap2, fmt='g', cbar=False)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
#Training and testing a decicion tree classifier - label
from sklearn.svm import SVC

clf1 = SVC(kernel='linear', C=1.0,random_state=76).fit(X1_train, y1_train)
y1_pred=clf1.predict(X1_test)

# Model Accuracy, how often is the classifier correct?
print("transport mode")
print("Accuracy:",metrics.accuracy_score(y1_test, y1_pred))
print("F1-Score:",metrics.f1_score(y1_test, y1_pred, average='macro'))
print("Confusion Matrix:\n", metrics.confusion_matrix(y1_test, y1_pred))


clf2 = SVC(kernel='linear', C=1.0,random_state=76).fit(X2_train, y2_train)
y2_pred=clf2.predict(X2_test)
print("label")
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y2_test, y2_pred))
print("F1-Score:",metrics.f1_score(y2_test, y2_pred, average='macro'))
print("Confusion Matrix:\n", metrics.confusion_matrix(y2_test, y2_pred))


```
